```python
from skimage import filters, color, feature, morphology, io
from matplotlib import pyplot as plt

image_list = ["samolot01.jpg","samolot07.jpg","samolot08.jpg","samolot17.jpg","samolot16.jpg","samolot09.jpg"]
images = []

for image in image_list:
  images.append(io.imread(image))

fig = plt.figure(facecolor="black", figsize=(14.0, 9.0))
i = 0

for image in images:
  plt.subplot(231+i)
  frame = plt.gca()
  frame.axes.get_xaxis().set_visible(False)
  frame.axes.get_yaxis().set_visible(False)
  image = color.rgb2gray(image)
  image = filters.sobel(image**3.2)
  image = feature.canny(image, sigma=1.1)
  #image = morphology.erosion(image, morphology.square(1))
  image = morphology.dilation(image, morphology.square(3))
  plt.imshow(image, cmap='gray')
  i += 1
plt.tight_layout()
plt.show()
#fig.savefig("samoloty.pdf",facecolor="black")
plt.close()
```
